```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.patches import Rectangle

# Set modern style
plt.style.use('default')
sns.set_palette("Set2")

# Custom color palette
colors = {
    'adaptive': '#3B82F6',    # Beautiful blue
    'baseline': '#EF4444',    # Beautiful red
    'meta': '#10B981',        # Green
    'tsla': '#F59E0B',        # Orange
    'nvidia': '#8B5CF6',      # Purple
    'background': '#F8FAFC',  # Light gray
    'grid': '#E2E8F0'         # Grid color
}

# Data preparation
meta_data = {
    'Model': ['RNN', 'LSTM', 'GRU', 'LINEAR'],
    'Adaptive_RMSE': [41.262, 41.990, 39.454, 58.939],
    'Baseline_RMSE': [71.625, 67.935, 60.279, 68.730],
    'Adaptive_MAE': [38.079, 40.817, 38.042, 53.916],
    'Baseline_MAE': [59.602, 55.388, 49.101, 57.124]
}

tsla_data = {
    'Model': ['RNN', 'LSTM', 'GRU', 'LINEAR'],
    'Adaptive_RMSE': [33.037, 30.729, 21.622, 25.923],
    'Baseline_RMSE': [56.658, 55.241, 48.494, 58.276],
    'Adaptive_MAE': [29.311, 28.605, 19.732, 20.904],
    'Baseline_MAE': [47.896, 46.765, 40.507, 47.879]
}

nvidia_data = {
    'Model': ['RNN', 'LSTM', 'GRU', 'LINEAR'],
    'Adaptive_RMSE': [6.256, 7.542, 6.138, 8.009],
    'Baseline_RMSE': [17.264, 16.717, 16.260, 9.935],
    'Adaptive_MAE': [5.971, 7.411, 5.896, 7.315],
    'Baseline_MAE': [14.545, 14.178, 13.639, 7.948]
}

# Create DataFrames
meta_df = pd.DataFrame(meta_data)
tsla_df = pd.DataFrame(tsla_data)
nvidia_df = pd.DataFrame(nvidia_data)

def create_beautiful_bar_chart(data_adaptive, data_baseline, title, ylabel, stock_color):
    """Create a beautiful bar chart with modern styling"""
    fig, ax = plt.subplots(figsize=(12, 8))
    fig.patch.set_facecolor(colors['background'])
    ax.set_facecolor('white')
    
    x = np.arange(len(data_adaptive))
    width = 0.35
    
    # Create bars with rounded corners effect
    bars1 = ax.bar(x - width/2, data_adaptive, width, 
                   label='🔥 Adaptive CV (Better)', 
                   color=colors['adaptive'], alpha=0.9,
                   edgecolor='white', linewidth=2)
    
    bars2 = ax.bar(x + width/2, data_baseline, width, 
                   label='📊 Baseline CV', 
                   color=colors['baseline'], alpha=0.9,
                   edgecolor='white', linewidth=2)
    
    # Styling
    ax.set_xlabel('Machine Learning Models', fontsize=14, fontweight='bold', color='#374151')
    ax.set_ylabel(ylabel, fontsize=14, fontweight='bold', color='#374151')
    ax.set_title(title, fontsize=18, fontweight='bold', color='#1F2937', pad=20)
    ax.set_xticks(x)
    ax.set_xticklabels(['RNN', 'LSTM', 'GRU', 'LINEAR'], fontsize=12, fontweight='600')
    
    # Beautiful grid
    ax.grid(axis='y', alpha=0.3, linestyle='--', linewidth=1, color=colors['grid'])
    ax.set_axisbelow(True)
    
    # Legend styling
    legend = ax.legend(fontsize=12, loc='upper right', 
                      frameon=True, fancybox=True, shadow=True,
                      borderpad=1, columnspacing=2)
    legend.get_frame().set_facecolor('white')
    legend.get_frame().set_alpha(0.9)
    
    # Add value labels with better positioning
    for i, (v1, v2) in enumerate(zip(data_adaptive, data_baseline)):
        # Adaptive values
        ax.text(i - width/2, v1 + max(data_baseline) * 0.01, 
               f'{v1:.1f}' if v1 >= 10 else f'{v1:.2f}', 
               ha='center', va='bottom', fontsize=11, 
               fontweight='bold', color=colors['adaptive'])
        
        # Baseline values
        ax.text(i + width/2, v2 + max(data_baseline) * 0.01, 
               f'{v2:.1f}' if v2 >= 10 else f'{v2:.2f}', 
               ha='center', va='bottom', fontsize=11, 
               fontweight='bold', color=colors['baseline'])
        
        # Improvement percentage
        improvement = ((v2 - v1) / v2) * 100
        ax.text(i, max(data_baseline) * 0.95, 
               f'↓{improvement:.1f}%', 
               ha='center', va='top', fontsize=10, 
               fontweight='bold', color='#059669',
               bbox=dict(boxstyle="round,pad=0.3", facecolor='#D1FAE5', alpha=0.8))
    
    # Remove top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color('#D1D5DB')
    ax.spines['bottom'].set_color('#D1D5DB')
    
    plt.tight_layout()
    plt.show()

def create_comparison_chart(title, ylabel, meta_data, tsla_data, nvidia_data):
    """Create a beautiful comparison chart for all stocks"""
    fig, ax = plt.subplots(figsize=(14, 10))
    fig.patch.set_facecolor(colors['background'])
    ax.set_facecolor('white')
    
    models = ['RNN', 'LSTM', 'GRU', 'LINEAR']
    x_pos = np.arange(len(models))
    width = 0.25
    
    # Create bars
    bars1 = ax.bar(x_pos - width, meta_data, width, 
                   label='📈 META', color=colors['meta'], alpha=0.9,
                   edgecolor='white', linewidth=2)
    
    bars2 = ax.bar(x_pos, tsla_data, width, 
                   label='🚗 TESLA', color=colors['tsla'], alpha=0.9,
                   edgecolor='white', linewidth=2)
    
    bars3 = ax.bar(x_pos + width, nvidia_data, width, 
                   label='🎮 NVIDIA', color=colors['nvidia'], alpha=0.9,
                   edgecolor='white', linewidth=2)
    
    # Styling
    ax.set_xlabel('Machine Learning Models', fontsize=14, fontweight='bold', color='#374151')
    ax.set_ylabel(ylabel, fontsize=14, fontweight='bold', color='#374151')
    ax.set_title(title, fontsize=18, fontweight='bold', color='#1F2937', pad=20)
    ax.set_xticks(x_pos)
    ax.set_xticklabels(models, fontsize=12, fontweight='600')
    
    # Beautiful grid
    ax.grid(axis='y', alpha=0.3, linestyle='--', linewidth=1, color=colors['grid'])
    ax.set_axisbelow(True)
    
    # Legend
    legend = ax.legend(fontsize=12, loc='upper right', 
                      frameon=True, fancybox=True, shadow=True,
                      borderpad=1, columnspacing=2)
    legend.get_frame().set_facecolor('white')
    legend.get_frame().set_alpha(0.9)
    
    # Add value labels
    for i, (meta_val, tsla_val, nvidia_val) in enumerate(zip(meta_data, tsla_data, nvidia_data)):
        ax.text(i - width, meta_val + max(max(meta_data), max(tsla_data), max(nvidia_data)) * 0.01, 
               f'{meta_val:.1f}' if meta_val >= 10 else f'{meta_val:.2f}', 
               ha='center', va='bottom', fontsize=10, 
               fontweight='bold', color=colors['meta'])
        
        ax.text(i, tsla_val + max(max(meta_data), max(tsla_data), max(nvidia_data)) * 0.01, 
               f'{tsla_val:.1f}' if tsla_val >= 10 else f'{tsla_val:.2f}', 
               ha='center', va='bottom', fontsize=10, 
               fontweight='bold', color=colors['tsla'])
        
        ax.text(i + width, nvidia_val + max(max(meta_data), max(tsla_data), max(nvidia_data)) * 0.01, 
               f'{nvidia_val:.1f}' if nvidia_val >= 10 else f'{nvidia_val:.2f}', 
               ha='center', va='bottom', fontsize=10, 
               fontweight='bold', color=colors['nvidia'])
    
    # Remove spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color('#D1D5DB')
    ax.spines['bottom'].set_color('#D1D5DB')
    
    plt.tight_layout()
    plt.show()

# Generate all beautiful charts

# META Charts
print("🏢 META STOCK ANALYSIS")
print("=" * 50)
create_beautiful_bar_chart(
    meta_df['Adaptive_RMSE'], meta_df['Baseline_RMSE'],
    '🏢 META Stock - RMSE Performance Comparison',
    'Root Mean Square Error (RMSE)', colors['meta']
)

create_beautiful_bar_chart(
    meta_df['Adaptive_MAE'], meta_df['Baseline_MAE'],
    '🏢 META Stock - MAE Performance Comparison',
    'Mean Absolute Error (MAE)', colors['meta']
)

# TESLA Charts
print("\n🚗 TESLA STOCK ANALYSIS")
print("=" * 50)
create_beautiful_bar_chart(
    tsla_df['Adaptive_RMSE'], tsla_df['Baseline_RMSE'],
    '🚗 TESLA Stock - RMSE Performance Comparison',
    'Root Mean Square Error (RMSE)', colors['tsla']
)

create_beautiful_bar_chart(
    tsla_df['Adaptive_MAE'], tsla_df['Baseline_MAE'],
    '🚗 TESLA Stock - MAE Performance Comparison',
    'Mean Absolute Error (MAE)', colors['tsla']
)

# NVIDIA Charts
print("\n🎮 NVIDIA STOCK ANALYSIS")
print("=" * 50)
create_beautiful_bar_chart(
    nvidia_df['Adaptive_RMSE'], nvidia_df['Baseline_RMSE'],
    '🎮 NVIDIA Stock - RMSE Performance Comparison',
    'Root Mean Square Error (RMSE)', colors['nvidia']
)

create_beautiful_bar_chart(
    nvidia_df['Adaptive_MAE'], nvidia_df['Baseline_MAE'],
    '🎮 NVIDIA Stock - MAE Performance Comparison',
    'Mean Absolute Error (MAE)', colors['nvidia']
)

# Overall Comparison Charts
print("\n🔥 OVERALL PERFORMANCE COMPARISON")
print("=" * 50)
create_comparison_chart(
    '🏆 Best RMSE Performance - All Stocks Comparison (Adaptive CV)',
    'Root Mean Square Error (RMSE)',
    meta_df['Adaptive_RMSE'], tsla_df['Adaptive_RMSE'], nvidia_df['Adaptive_RMSE']
)

create_comparison_chart(
    '🏆 Best MAE Performance - All Stocks Comparison (Adaptive CV)',
    'Mean Absolute Error (MAE)',
    meta_df['Adaptive_MAE'], tsla_df['Adaptive_MAE'], nvidia_df['Adaptive_MAE']
)

# Beautiful Summary Table
print("\n" + "="*120)
print("📊 PERFORMANCE SUMMARY - ALL STOCKS")
print("="*120)

models = ['RNN', 'LSTM', 'GRU', 'LINEAR']
summary_data = {
    'Stock': ['META'] * 4 + ['TSLA'] * 4 + ['NVIDIA'] * 4,
    'Model': models * 3,
    'Adaptive_RMSE': list(meta_df['Adaptive_RMSE']) + list(tsla_df['Adaptive_RMSE']) + list(nvidia_df['Adaptive_RMSE']),
    'Adaptive_MAE': list(meta_df['Adaptive_MAE']) + list(tsla_df['Adaptive_MAE']) + list(nvidia_df['Adaptive_MAE']),
    'Baseline_RMSE': list(meta_df['Baseline_RMSE']) + list(tsla_df['Baseline_RMSE']) + list(nvidia_df['Baseline_RMSE']),
    'Baseline_MAE': list(meta_df['Baseline_MAE']) + list(tsla_df['Baseline_MAE']) + list(nvidia_df['Baseline_MAE']),
}

summary_df = pd.DataFrame(summary_data)
summary_df['RMSE_Improvement_%'] = ((summary_df['Baseline_RMSE'] - summary_df['Adaptive_RMSE']) / summary_df['Baseline_RMSE'] * 100).round(1)
summary_df['MAE_Improvement_%'] = ((summary_df['Baseline_MAE'] - summary_df['Adaptive_MAE']) / summary_df['Baseline_MAE'] * 100).round(1)

print(summary_df.to_string(index=False))

# Best Performing Models
print("\n" + "="*80)
print("🏆 CHAMPION MODELS FOR EACH STOCK")
print("="*80)

def find_best_model(df, metric_col):
    return df.loc[df[metric_col].idxmin(), 'Model']

print(f"🏢 META Stock Champions:")
print(f"   🥇 Best RMSE: {find_best_model(meta_df, 'Adaptive_RMSE')} ({min(meta_df['Adaptive_RMSE']):.3f})")
print(f"   🥇 Best MAE:  {find_best_model(meta_df, 'Adaptive_MAE')} ({min(meta_df['Adaptive_MAE']):.3f})")

print(f"\n🚗 TESLA Stock Champions:")
print(f"   🥇 Best RMSE: {find_best_model(tsla_df, 'Adaptive_RMSE')} ({min(tsla_df['Adaptive_RMSE']):.3f})")
print(f"   🥇 Best MAE:  {find_best_model(tsla_df, 'Adaptive_MAE')} ({min(tsla_df['Adaptive_MAE']):.3f})")

print(f"\n🎮 NVIDIA Stock Champions:")
print(f"   🥇 Best RMSE: {find_best_model(nvidia_df, 'Adaptive_RMSE')} ({min(nvidia_df['Adaptive_RMSE']):.3f})")
print(f"   🥇 Best MAE:  {find_best_model(nvidia_df, 'Adaptive_MAE')} ({min(nvidia_df['Adaptive_MAE']):.3f})")

# NVIDIA Detailed Analysis
print("\n" + "="*80)
print("🎮 NVIDIA DETAILED PERFORMANCE ANALYSIS")
print("="*80)
print(f"{'Model':<8} {'Adaptive':<12} {'Baseline':<12} {'Improvement':<12} {'% Better':<12}")
print("=" * 80)

for i, model in enumerate(models):
    adaptive_rmse = nvidia_df['Adaptive_RMSE'].iloc[i]
    baseline_rmse = nvidia_df['Baseline_RMSE'].iloc[i]
    improvement = baseline_rmse - adaptive_rmse
    pct_improvement = (improvement / baseline_rmse) * 100
    print(f"{model:<8} {adaptive_rmse:<12.3f} {baseline_rmse:<12.3f} {improvement:<12.3f} ↓{pct_improvement:<11.1f}%")

print(f"\n{'Model':<8} {'Adaptive':<12} {'Baseline':<12} {'Improvement':<12} {'% Better':<12}")
print("=" * 80)
for i, model in enumerate(models):
    adaptive_mae = nvidia_df['Adaptive_MAE'].iloc[i]
    baseline_mae = nvidia_df['Baseline_MAE'].iloc[i]
    improvement = baseline_mae - adaptive_mae
    pct_improvement = (improvement / baseline_mae) * 100
    print(f"{model:<8} {adaptive_mae:<12.3f} {baseline_mae:<12.3f} {improvement:<12.3f} ↓{pct_improvement:<11.1f}%")

print("\n🚀 KEY INSIGHTS:")
print("=" * 50)
print("• 🎯 NVIDIA shows the LOWEST prediction errors across all models")
print("• 🔥 GRU model performs BEST for both NVIDIA RMSE and MAE")
print("• 💡 Adaptive CV consistently OUTPERFORMS Baseline CV")
print("• 📈 NVIDIA improvements range from 19.4% to 63.7% with Adaptive CV")
print("• 🏆 All stocks benefit significantly from Adaptive Cross-Validation")
```
